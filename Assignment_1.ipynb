## Objective: - Predict Next Sequence
To start with deep learning, the very basic project that you can build is to
predict the next digit in a sequence.

Dataset: - Create a sequence like a list of odd numbers and then build a model
and train it to predict the next digit in the sequence. bold text

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import tensorflow as tf
import seaborn as sns
```

```python
# define the keras model

model_clf = tf.keras.Sequential( 
    [
          tf.keras.layers.Dense(units =2,input_shape=[1]),
          tf.keras.layers.Dense(units =1)
    ]
    )

```

```python
model_clf.layers
```

```python
LOSS_FUNCTION = "mean_squared_error" # use => tf.losses.sparse_categorical_crossentropy
OPTIMIZER = "SGD" # or use with custom learning rate=> tf.keras.optimizers.SGD(0.02)


model_clf.compile(loss=LOSS_FUNCTION,
              optimizer=OPTIMIZER,
             )
```

```python
xs = np.array([1,2,3,5],dtype=float)
ys = np.array([21,23,25,27],dtype=float)
```

```python
EPOCHS = 2000
history = model_clf.fit(xs, ys, epochs=EPOCHS)
```

```python
results = model_clf.predict([6])
rounded_results = [int(np.round(num)) for num in results]
print(rounded_results)
```
